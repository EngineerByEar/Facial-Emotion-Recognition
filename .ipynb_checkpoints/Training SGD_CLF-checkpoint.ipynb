```python
import pandas as pd
import numpy as np
from PIL import Image
import os
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import RandomizedSearchCV
import pickle

from data_load_utils import load_image, load_data, create_data_frame
```

# Loading the Data

```python
training_dir = "./data/train"
testing_dir = "./data/test"
```

```python
emotions, images = load_data(training_dir)
le = LabelEncoder()
labels = le.fit_transform(emotions)
```

```python
training_data = create_data_frame(emotions, labels, images)
```

```python
test_emotions, test_images = load_data(testing_dir)
test_labels = le.transform(test_emotions)
```

```python
testing_data = create_data_frame(test_emotions, test_labels, test_images)
```

## Preparing Data for SGD Model

```python
def prepare_data_sgd(df):
    def flatten_and_scale(img):
        return img.flatten()/255
    df["image"] = df["image"].apply(flatten_and_scale)
    X = np.stack(df["image"].values)
    y = df["label"].values
    return X,y
```

```python
X_train, y_train = prepare_data_sgd(training_data)
```

```python
X_test, y_test = prepare_data_sgd(testing_data)
```

## Finding best Model Params with RandomizedSearchCV

```python
parameters = {
    "loss": ["hinge", "log_loss"],
    "alpha": [0.0001, 0.001, 0.01],
    "penalty": ["l2", "l1"],
    "learning_rate": ["optimal", "adaptive"],
}
```

```python
sgd = SGDClassifier()
```

```python
clf = RandomizedSearchCV(sgd, parameters, n_iter = 10, cv = 2, verbose = 3)
```

```python
clf.fit(X_train, y_train)
```

```python
best_params = clf.best_params_
```

```python
print(best_params)
```

## RandomizedSearchCV graph

```python
scores = clf.cv_results_["split1_test_score"]
```

```python
scores.sort()
```

```python
plt.bar(range(len(scores)), scores)
plt.title("Accuracy Scores in RandomizedSearchCV on SGD-Classifier")
plt.ylabel("Accuracy Score")
plt.xlabel("Model")
plt.xticks(range(len(scores)))
plt.ylim(0, 0.4)
for i, score in enumerate(scores):
    plt.text(i-0.42, score+ 0.005, f"{score:.3f}")
plt.savefig("./graphics/SGD_RandomizedSearchCV")
plt.show()
```

## Training Model on Best Params

```python
best_sgd = SGDClassifier(penalty = "l2", loss = "log_loss", learning_rate = "adaptive", alpha = 0.01)
```

```python
best_sgd.fit(X_train, y_train)
```

```python
#Saving Model
with open("./models/sgd.pickle", "wb") as f:
    pickle.dump(best_sgd, f)
```

```python
prediction = best_sgd.predict(X_test)
```

```python
acc = accuracy_score(y_test, prediction)
```

```python
print(f"Accuracy Score: {acc}")
```
